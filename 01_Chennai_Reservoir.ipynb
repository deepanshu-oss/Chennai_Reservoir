### Inferences

* Although Chennai receive some rainfall in July , Aug , Sep due to South-West monsoon but majority of the rainfall happens in Nov-Dec due North-East monsoon
* From 2006 to 2010, North East monsoon had much more contribution than South-West monsoon. In last few years , however, either both SW and NE are contributing equally or SW is contributing more.
* Declining in NE monsoon rainfall in second half of 2010s
